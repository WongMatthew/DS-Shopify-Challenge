```python
'''
Need to figure out WHAT went wrong when evaluating the AOV

Noticed that there are 5000 rows, so decided to take the total order amount / total rows 
to see if that gives an AOV of 3145.128
'''

print('AOV:',data['order_amount'].sum()/5000)
```

```python
'''
while $3145.128 is technically not wrong because you're dividing by the total num of orders, the orders fluctuate in size

the issue with dividing by 5000 is that each order can consist of multiple products, take a look at row 15

the order amount (value of the order) is HUGE, under the previously calculated AOV, you'd assume that this row is a single 
product, however, it is actually 2000 products in a single order
'''

print(data.loc[[15]])
```

```python
'''
total AOV is normally calculated like: AOV = total revenue / total orders

I think it would be more optimal to calculate the average revenue per item!

by taking the sum of 'total items' we can appropriately value the average revenue from each item 
'''
print('Avg Revenue per Item:',data['order_amount'].sum()/data['total_items'].sum())
```
